```python
import numpy as np
import matplotlib.pyplot as plt

def compare_waveforms(tnodiff, nodiff, tdiff, diff, label, pdfname):
    fig, ax = plt.subplots(len(tnodiff), 2, figsize=(5*2, len(tnodiff)*4))
    for i, (time_nodiff, wf_nodiff, time_diff, wf_diff, name) in enumerate(zip(tnodiff, nodiff, tdiff, diff, pdfname)):

        ax[i, 0].plot(time_nodiff, wf_nodiff, label='w/o diffusion')
        ax[i, 0].plot(time_diff, wf_diff, '--', label='w/ diffusion')
        ax[i, 1].plot(time_diff, (np.array(wf_diff) - np.array(wf_nodiff)) / np.array(wf_nodiff) )

        ax[i, 0].legend(prop={'size':12})
        ax[i, 0].set_xlabel('drift time [us]', fontsize=12)
        ax[i, 0].set_ylabel(label, fontsize=12)
        ax[i, 0].tick_params(axis='both', labelsize=12)
        ax[i, 1].set_xlabel('drift time [us]', fontsize=12)
        ax[i, 1].set_ylabel('relative difference', fontsize=12)
        ax[i, 1].tick_params(axis='both', labelsize=12)
        ax[i, 0].set_title(name, fontsize=12)
        #ax[i, 0].semilogy()
    plt.tight_layout()
```

```python
def zposition_effect_onDiffusedWF(tnodiff, nodiff, tdiff, diff, label, pdfname):
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    for t0, wf0, t1, wf1, name in zip(tnodiff, nodiff, tdiff, diff, pdfname):
        ax[0].plot(range(len(wf0)), wf0, ':',  label=name+' w/o diffusion')
        ax[0].plot(range(len(wf1)), wf1, '-.', label=name+' w/  diffusion')
        
    ax[0].legend(prop={'size':10}, loc='upper left')
    ax[0].set_xlabel('shifted time [us]', fontsize=12)
    ax[0].set_ylabel('current amplitude', fontsize=12)
    ax[0].tick_params(axis='both', labelsize=12)
    
    x, rel_diff = [], []
    for i, (v1, v2) in enumerate(zip(diff[0], diff[1])):
        if v1 != 0 and v2 != 0:
            x.append(i)
            rel_diff.append( (v1-v2)/v1*100)
    ax[1].plot(x, rel_diff, color='black')
    ax[1].tick_params(axis='both', labelsize=12)
    ax[1].set_xlabel('shifted time [us]', fontsize=12)
    ax[1].set_ylabel('relative difference [%]', fontsize=12)
    plt.tight_layout()
    return fig
```

```python
# Try to decide the diffusion PDF scanning grid granularity.
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from scipy.interpolate import griddata

def uniform_sample_fixed_sum(size, target_num, low, up):
    q_depo = np.random.uniform(low, up, size=size)
    scaled_q_depo = q_depo / np.sum(q_depo) * target_num
    return scaled_q_depo

def normal_sample_fixed_sum(size, target_num, loc, scale):
    q_depo = np.random.normal(loc=loc, scale=scale, size=size)
    scaled_q_depo = q_depo / np.sum(q_depo) * target_num
    return scaled_q_depo


n_iter = 5000
gen_wfs = []

xmin, xmax, ymin, ymax = 2, 3, 0, 1


for i_iter in tqdm(range(n_iter)):
    n_depo = 10
    x_depo = np.random.uniform( xmin, xmax, size=n_depo)
    y_depo = np.random.uniform(ymin, ymax, size=n_depo)
    #q_depo = np.random.uniform(0., 2.0, size=n_depo)
    q_depo = uniform_sample_fixed_sum(n_depo, 1, 0, 1.0)
    #q_depo = normal_sample_fixed_sum(n_depo, 1, 1, 0.1)

    wfs = []
    filelist = []
    for i_depo in range(n_depo):
        x_pdf_low = int(3 - np.abs((x_depo[i_depo])%6 - 3))
        y_pdf_low = int(np.abs(y_depo[i_depo]))
        x_pdf_high, y_pdf_high = x_pdf_low + 1, y_pdf_low + 1 
        
        pdf_filename_xlowylow = f'/Users/yumiao/Documents/Works/0nbb/nEXO/Reconstruction/waveform/nEXO_reconstruction/PDFs/diffusionPDFs_YstripatX0.0Y0.0_chargeX{x_pdf_low:.1f}Y{y_pdf_low:.1f}Z619.63Q100000.0.npy'
        pdf_filename_xlowyhigh = f'/Users/yumiao/Documents/Works/0nbb/nEXO/Reconstruction/waveform/nEXO_reconstruction/PDFs/diffusionPDFs_YstripatX0.0Y0.0_chargeX{x_pdf_low:.1f}Y{y_pdf_high:.1f}Z619.63Q100000.0.npy'
        pdf_filename_xhighylow = f'/Users/yumiao/Documents/Works/0nbb/nEXO/Reconstruction/waveform/nEXO_reconstruction/PDFs/diffusionPDFs_YstripatX0.0Y0.0_chargeX{x_pdf_high:.1f}Y{y_pdf_low:.1f}Z619.63Q100000.0.npy'
        pdf_filename_xhighyhigh = f'/Users/yumiao/Documents/Works/0nbb/nEXO/Reconstruction/waveform/nEXO_reconstruction/PDFs/diffusionPDFs_YstripatX0.0Y0.0_chargeX{x_pdf_high:.1f}Y{y_pdf_high:.1f}Z619.63Q100000.0.npy'
        

        if not os.path.exists(pdf_filename_xlowylow):
            print(f'Error: PDF {pdf_filename_xlowylow} does not exist :( !!!!!')
        else:
            t0_xlowylow, wf0_xlowylow = np.load(pdf_filename_xlowylow)

        if not os.path.exists(pdf_filename_xlowyhigh):
            print(f'Error: PDF {pdf_filename_xlowyhigh} does not exist :( !!!!!')
        else:
            t0_xlowyhigh, wf0_xlowyhigh = np.load(pdf_filename_xlowyhigh)


        if not os.path.exists(pdf_filename_xhighylow):
            print(f'Error: PDF {pdf_filename_xhighylow} does not exist :( !!!!!')
        else:
            t0_xhighylow, wf0_xhighylow = np.load(pdf_filename_xhighylow)


        if not os.path.exists(pdf_filename_xhighyhigh):
            print(f'Error: PDF {pdf_filename_xhighyhigh} does not exist :( !!!!!')
        else:
            t0_xhighyhigh, wf0_xhighyhigh = np.load(pdf_filename_xhighyhigh)


        Xs = [xmin, xmin, xmax, xmax]
        Ys = [ymin, ymax, ymin, ymax]
        Zs = np.array([wf0_xlowylow, wf0_xlowyhigh, wf0_xhighylow, wf0_xhighyhigh])
        f = griddata((Xs, Ys), Zs, (x_depo[i_depo], y_depo[i_depo]), method='linear')

        wfs.append(f)                 


    wfs = np.array(wfs)
    wf_gen = np.sum(q_depo[:, np.newaxis] * wfs, axis=0)
    gen_wfs.append(wf_gen)


fig, ax = plt.subplots(figsize=(7, 5))
for i_iter in range(n_iter):
    ax.plot(t0_xhighyhigh[-60:-20], gen_wfs[i_iter][-60:-20], alpha=0.3)
ax.set_xlabel('time [us]', fontsize=12)
ax.set_ylabel('current amplitude', fontsize=12)
ax.tick_params(axis='both', labelsize=12)    
ax.set_title(f'X from {xmin} to {xmax} mm, Y from {ymin} to {ymax} mm', fontsize=13)
plt.tight_layout()
    
```
